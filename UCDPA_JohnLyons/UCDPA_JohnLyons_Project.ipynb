```python
# Replacing na values in Null Columns with proper explanation. Easier for another person to view and when thinking about constructing graphs easier to seperate data with distinct names
```

```python
Data_Sal_Jobs_df["work_year"].fillna("0", inplace = True) 
Data_Sal_Jobs_df["experience_level"].fillna("Experience Unknown", inplace = True) 
Data_Sal_Jobs_df["employment_type"].fillna("Type of Employment Unknown", inplace = True) 
Data_Sal_Jobs_df["job_title"].fillna("Job Title Unknown", inplace = True) 
Data_Sal_Jobs_df["salary"].fillna("0.0", inplace = True) 
Data_Sal_Jobs_df["salary_currency"].fillna("Salary Currency Unknown", inplace = True) 
Data_Sal_Jobs_df["salary_in_usd"].fillna("0.0", inplace = True) 
Data_Sal_Jobs_df["employee_residence"].fillna("Employee Residence Unknown", inplace = True) 
Data_Sal_Jobs_df["remote_ratio"].fillna("0.0", inplace = True) 
Data_Sal_Jobs_df["company_location"].fillna("Employee's Company Location", inplace = True) 
Data_Sal_Jobs_df["company_size"].fillna("Size of Employee's Company Unknown", inplace = True) 
Data_Sal_Jobs_df["company_name"].fillna("Name of Employee's Company Unknown", inplace = True) 
Data_Sal_Jobs_df["title"].fillna("Employee's Title in their Company Unknown", inplace = True) 
Data_Sal_Jobs_df["remote"].fillna("Unknown if Employee works Remotely", inplace = True) 
Data_Sal_Jobs_df["url"].fillna("Company URL Unknown", inplace = True) 
Data_Sal_Jobs_df["tags"].fillna("Tags associated with Employee's Role Unknown", inplace = True) 
Data_Sal_Jobs_df["job_types"].fillna("Job Type associated with Employee's Role Unknown", inplace = True)
Data_Sal_Jobs_df["location"].fillna("Company's Location Unknown", inplace = True)
```
